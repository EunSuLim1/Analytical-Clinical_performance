```python
df1 = pd.read_excel("C:/Users/pcl/Roma_raw/211021 CT211018 cal.xlsx")
df2 = df1.iloc[:,1:]
df2
```

```python
EX3 = df2['Exposed time3'].values.tolist()
EX1 = df2['Exposed time1'].values.tolist()

ca125_raw=[]
HE4_raw=[]
for i in df2.index.tolist():
    if df2['Marker'].iloc[i] == 'CA125':
        ca125_raw.append(EX3[i])
    else:
        HE4_raw.append(EX1[i])
        
df_ca125 = pd.DataFrame(ca125_raw, columns=['s/g'])
df_ca125['conc'] = [2000,800,400,200,100,30]*3
df_ca125.astype('float')
df_ca125=df_ca125[['conc','s/g']]
df_ca125
```

```python
def logistic4 (x, A, B, C, D):
    return ((A-D)/(1+((x/C**B))))+D
```

```python
xdata = df_ca125['conc']
ydata = df_ca125['s/g']
popt, pcov = curve_fit(logistic4, xdata, ydata)
print(popt)
```

```python
x_fit = np.linspace(0.01,2000,2000) 
# x= np.linspace(start, stop, num) 시작, 끝, 개수 1차 배열
y_fit = logistic4(x_fit,*popt)
y_fit
```

```python
plt.plot(xdata, ydata, 'o', label = 'data')
plt.plot(x_fit, y_fit, label = 'fit')
plt.xscale('log')
```
